```python
import ipywidgets as widgets
from IPython.display import display
import  os
```

```python
def exponent(maxIteration):
    x = np.arange(5* maxIteration)
    y = np.exp(x)
    
    plt.scatter(x,y,color='g',marker='o')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title("widget")
    plt.show()
    
maxIteration = widgets.boundedIntText(min=1,max=1000,value=1,description='Iter')
widgets.interactive(exponent,maxIteration=maxIteration)
```
